# Plotting Integrals for Mathematical Methods Problem Sheet One

## Q4.1

```python
def f_integrand(x,p):
    numerator = np.e**(-p*x**2)
    denominator = 1 + x
    return numerator/denominator
    
    
def reimann_f(f,p):
    '''
    p is array!
    '''
    start_time = time()
    print(start_time)
    a = 0
    b = 100
    n = 10000000
    
    f_values = []
    for i in p:   
    
        x = np.linspace(a,b,n)
        p_i = np.linspace(i,i,n)
    
        delta_x = x[1]-x[0]
        f_approx = f(x,p_i)
    
        f_approx = f_approx[0:-1]
        x = x[0:-1]
    
        # calculate integral value
        integral = sum(f_approx)*delta_x
        f_values.append(integral)
        
    end_time = time()
    elapsed = end_time - start_time
    print('Time taken to solve numerically using', n, 'reimann slices for 100 integrals of p:')
    print(round(elapsed,4),'s')
    return f_values
    
def ff(p):
    return (1/2)*(np.pi/p)**(1/2) - 1/(2*p)
```

```python
p = np.linspace(0.5,10,100)
f_compare = reimann_f(f_integrand,p)
```

```python
plt.figure(figsize=[10,6])
plt.title('Solving integrals as functions: MM Problem Set 1 Q4.1')
plt.ylabel('y')
plt.xlabel('p')
plt.plot(p,f_compare,color='blue',label='y = F(p) integral function solved numerically')
plt.plot(p,ff(p),color='black',linestyle=':',label='y = FF(p) solved with approximation (wake up rule!)')
plt.legend(loc='upper right')
```

With approximation is LESS than true <br>
Expected as $(1+x)^{-1} > 1 - x$ 

## Q4.2

```python
def f_integrand(x):
    k = 1
    numerator = np.sin(k*x)
    denominator = x
    return numerator/denominator
    
    
def reimann_f(f,p):
    ''' p is array!'''
    start_time = time()
    print(start_time)
    a = 0.0001
    n = 10000
    
    f_values = []
    for i in p:   
    
        x = np.linspace(a,i,n)
    
        delta_x = x[1]-x[0]
        f_approx = f(x)
    
        f_approx = f_approx[0:-1]
        x = x[0:-1]
    
        # calculate integral value
        integral = sum(f_approx)*delta_x
        f_values.append(integral)
        
    end_time = time()
    elapsed = end_time - start_time
    print('Time taken to solve numerically using', n, 'reimann slices for 10000 integrals of p:')
    print(round(elapsed,4),'s')
    return f_values
    
def ff(p):
    return (1)*(p) - (1**3/18)*(p**3)
```

```python
p = np.linspace(0.0000001,2*np.pi,10000)
f_compare = reimann_f(f_integrand,p)
```

```python
plt.figure(figsize=[10,6])
plt.title('Solving integrals as functions: MM Problem Set 1 Q4.2')
plt.ylabel('y')
plt.xlabel('p')
plt.plot(p,f_compare,color='blue',label='y = F(p) integral function solved numerically')
plt.plot(p,ff(p),color='black',linestyle=':',label='y = FF(p) solved with approximation (small angle)')
plt.legend(loc='lower center')
```
